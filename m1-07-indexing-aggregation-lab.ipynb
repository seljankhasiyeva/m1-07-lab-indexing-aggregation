```python
# 1
import pandas as pd
scores=pd.read_csv("data_safe_copy.csv")
scores.head()
# scores.info()
# len(scores)
```

```python
# 2
roster=pd.DataFrame({
    "student_id":["S071", "S072", "S073", "S074", "S075", "S076", "S077", "S078", "S079", "S080", "S081", "S082", "S083", "S084"],
    "status": ["active", "inactive", "active", "inactive", "active", "active", "active", "inactive", "inactive", "active", "active", "inactive", "active", "inactive"]
})
merged=pd.merge(scores, roster, how="left")
merged
missing_statuses=round(merged["status"].isna().sum())
missing_statuses
```

```python
# 3 
import numpy as np
avg_by_module=scores.groupby(["cohort", "module"])["score"].mean().reset_index()
avg_by_cohort=scores.groupby(["cohort"])["score"].mean().reset_index()
check_roll_up=avg_by_module.groupby(["cohort"])["score"].mean()
comparison=np.allclose(avg_by_cohort["score"].values, check_roll_up.values)
print(f"Module-level averages roll up to Cohort-level averages: {comparison}")
```

```python
# 4
student_module_report=scores.pivot_table(
    index="student_id",
    columns="module",
    values="score",
    aggfunc="mean"
)
student_module_report
print(f"The number of rows equals the number of unique students in the original dataset: {len(student_module_report)==scores["student_id"].nunique()}")
print(f"The column labels match the set of module names: {list(student_module_report.columns)==sorted(scores["module"].unique().tolist())}")
```

```python
# 5
students_avg=scores.groupby(["student_id","cohort"])["score"].mean().reset_index()
top_students = scores.groupby(['student_id', 'cohort'])['score'].mean().reset_index().sort_values('score', ascending=False).groupby('cohort').head(3).sort_values(['cohort', 'score'], ascending=[True, False])
top_students = top_students.rename(columns={'score': 'avg_score'})
check_top = (top_students['cohort'].value_counts() == 3).all()
top_students
print(f"Each cohort appears exactly three times: {check_top}")
```
